**xi(i=1,2,3) = {1 if investment i is made | 0 otherwise}**  

**max z= 16x1 + 22x2 +12x3**  
        **S.t.**       
    **5x1 + 7x2 + 4x3 <= 10**  

```python
import numpy as np
from geneticalgorithm import geneticalgorithm as ga
```

```python
# Defining the equations 

def fitness_function(x):
    x1 = x[0]
    x2 = x[1]
    x3 = x[2]

    # Apply constraints
    penalty = 0
    if 5*x1 + 7*x2 + 4*x3 > 10:
        penalty = np.inf

    return -(16*x1 + 22*x2 + 12*x3) + penalty #objective function is negated because of maximization

# creating an instance of the GA solver

model = ga(function = fitness_function,dimension=3,variable_type='bool')

model.run()
    
```
